```python
import numpy as np
import pandas as pd
```

```python
import os
import sys
import git
import pathlib

PROJ_ROOT_PATH = pathlib.Path(git.Repo('.', search_parent_directories=True).working_tree_dir)
PROJ_ROOT =  str(PROJ_ROOT_PATH)
if PROJ_ROOT not in sys.path:
    sys.path.append(PROJ_ROOT)
    
from libs.constants import model_seeds, error_seeds
```

```python
model_arch = "mnist32-cnn_1024_256_64"
```

```python
# Load dictionary
results_tag = 'fault_injection_analysis_results-all_model_seeds--lowerror--'+model_arch
qent_results = np.load(results_tag + '.npy', allow_pickle='TRUE').item()
```

```python
pd.options.display.float_format = '{:.2%}'.format
```

```python
df = pd.DataFrame.from_dict({(i,j,k): qent_results[i][j][k] 
                           for i in list(qent_results.keys()) 
                           for j in list(qent_results[i].keys())[1:]
                           for k in list(qent_results[i][j].keys())},
                       orient='index')
df.index.names = ['model_seed','error_profile','error_type']
```

```python
df = df.unstack()
```

```python
df.to_csv(results_tag + '.csv')
```
